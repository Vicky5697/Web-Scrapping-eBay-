```python
import csv
import requests
from bs4 import BeautifulSoup
from decimal import Decimal
```

```python
# creating a varibale website and getting url request in the form of text.
website = requests.get('https://www.ebay.com/sch/i.html?_nkw=watch').text
```

```python
soup = BeautifulSoup(website, 'html.parser')
```

```python
items = soup.select('.srp-results .s-item')

# we get the item details here
for item in items:
    print(item.text)
```

```python
items = soup.select('.srp-results .s-item')

# Here we get the title of the product or we can say product name 
for item in items:
    title = item.select_one('h3').text
    print(title)
```

```python
items = soup.select('.srp-results .s-item')

# Here we get the title of the product or we can say product name with price 
for item in items:
    title = item.select_one('h3').text
    price = item.select_one('.s-item__price').text
    print(title)
    print(price)
    
```

```python
items = soup.select('.srp-results .s-item')

# Here we get the title of the product or we can say product name with price 
for item in items:
    title = item.select_one('h3').text
    price = item.select_one('.s-item__price').text
    ship = item.select_one('.s-item__shipping').text
    print(title)
    print(price)
    print(ship)
```

```python
# here we added a for loop to get all the title and price of the product from page 1 to 5.
for i in range(1,5):
    website = requests.get('https://www.ebay.com/sch/i.html?_nkw=watch&_pgn=', i).text
    soup = BeautifulSoup(website, 'html.parser')
    items = soup.select('.srp-results .s-item')
    for item in items:
        title = item.select_one('h3').text
        price = item.select_one('.s-item__price').text
        ship = item.select_one('.s-item__shipping').text
        print(title)
        print(price)
        print(ship)
```

```python
#Here we take user input for product and shipping and let it search for the product that the user wants 
query = input('Enter Product : ')
shipping = input('Enter Free Shipping : ')
for i in range(1,5):
    website = requests.get('https://www.ebay.com/sch/i.html?_nkw='+ query +'&_pgn=', i).text
    soup = BeautifulSoup(website, 'html.parser')
    items = soup.select('.srp-results .s-item')
    for item in items:
        title = item.select_one('h3').text
        price = item.select_one('.s-item__price').text
        ship = item.select_one('.s-item__shipping').text
        if shipping == 'y':
            if 'Free' in ship:
                print(title, price, ship)
                print('******************')
        else:
            print(title, price, ship)
            print('******************')
```

```python
query = input('Enter Product : ')
shipping = input('Enter Free Shipping : ')
max_price = Decimal(input('Enter Maximum price : '))
for i in range(1,5):
    website = requests.get('https://www.ebay.com/sch/i.html?_nkw='+ query +'&_pgn=', i).text
    soup = BeautifulSoup(website, 'html.parser')
    items = soup.select('.srp-results .s-item')
    for item in items:
        title = item.select_one('h3').text
        price = item.select_one('.s-item__price').text
        ship = item.select_one('.s-item__shipping').text
        price_decimal = Decimal(price.split(' to ')[0][1:])
        if price_decimal <= max_price:
            if shipping == 'y':
                if 'Free' in ship:
                    print(title,'\n', price,'\n', ship)
                    print('******************')
            else:
                print(title,'\n', price,'\n', ship)
                print('******************')
```
